# Basic Visualization

1. line chart : untuk lihat trend sebuah data
2. bar chart : untuk lihat komparasi
3. area chart : untuk melihat perubahan data terjadi seiring dengan waktu berjalan
4. pie chart : untuk melihat persentase
5. histogram : mengihtung distribusi frekuensi


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
pd.set_option('display.max_columns', None)  # Menampilkan semua kolom
pd.set_option('display.max_rows', None)     # Menampilkan semua baris
pd.set_option('display.max_colwidth', None) # Menampilkan semua lebar kolom
```

```python
#line chart
data = {
    'tahun' : ['2010', '2011', '2012', '2013','2014', '2015', '2016', '2017', '2018', '2019', '2020'],
    'penjualan' : [10, 12, 15,20, 18,25,30,35,40,45,50]
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
plt.plot(df['tahun'], df['penjualan'])
plt.xlabel('Tahun')
plt.ylabel('Penjualan')
plt.title('Grafik penjualan per Tahun')
plt.show()
```

```python
#bar chart
plt.bar(df['tahun'], df['penjualan'])
plt.xlabel('Tahun')
plt.ylabel('Penjualan')
plt.title('Grafik penjualan per Tahun')
plt.show()
```

```python
#area chart
plt.fill_between(df['tahun'], df['penjualan'], color='skyblue')
plt.plot(df['tahun'], df['penjualan'], color='blue')
plt.xlabel('Tahun')
plt.ylabel('Penjualan')
plt.title('Grafik penjualan per Tahun')
plt.show()
```

```python
#pie chart
data_pie = {
    'kategori' : ['A', 'B', 'C', 'D', 'E'],
    'nilai' : [20, 10,15,5,25]
}
df = pd.DataFrame(data_pie)
df.head()
```

```python
plt.pie(data_pie['nilai'], labels=data_pie['kategori'], autopct='%1.1f%%')
plt.title('persentase kategori')
plt.show()
```

```python
#histogram
#bins => pembagiannya berapa ?

data_hist = {
    'nilai' :[55,60,70,75,80,82,85,90,95,100]
}

df = pd.DataFrame(data_hist)
df.head()
```

```python
plt.hist(df['nilai'], bins=5)
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('distribusi nilai')
plt.show()
```
