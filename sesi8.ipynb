### Advance Visualization

```python
#importing library
from wordcloud import WordCloud, STOPWORDS
import PyPDF2 #untuk baca text dari PDF
import matplotlib.pyplot as plt
```

```python
#function untuk membaca text dari file pdf
def extract_text_from_pdf(pdf_file) :
    with open(pdf_file, 'rb') as pdf_reader :
        reader = PyPDF2.PdfReader(pdf_reader)
        text = '' #untuk tampung text dari pdf nya
        for page in reader.pages :
            text += page.extract_text() #extract text & concate ke text
    return text
```

```python
#load pdf file 
pdf_file = 'data/D9439118419.pdf'
text_data = extract_text_from_pdf(pdf_file)
```

```python
wordcloud = WordCloud(
    background_color='white',
    width=800,
    height=800,
).generate(text_data)
```

```python
fig = plt.figure()
fig.set_figheight(14)
fig.set_figheight(18)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords = set(STOPWORDS)
stopwords.add('using')
stopwords.add('used')
wordcloud = WordCloud(
    background_color='white',
    width=800,
    height=800,
    stopwords=stopwords
).generate(text_data)
```

```python
fig = plt.figure()
fig.set_figheight(14)
fig.set_figheight(18)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
